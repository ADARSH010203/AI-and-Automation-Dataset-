```python
import matplotlib.pyplot as plt

# Clean data for analysis
a_cleaned = a.copy()

# Convert percentage columns to numeric by removing % and converting to float
percentage_columns = [
    'AI Adoption (%)', 'Global Expectation for AI Adoption (%)',
    'Estimated Jobs Eliminated by AI (millions)', 'Estimated New Jobs Created by AI (millions)', 
    'Net Job Loss in the US', 'Organizations Believing AI Provides Competitive Edge',
    'Companies Prioritizing AI in Strategy', 'Marketers Believing AI Improves Email Revenue',
    'Expected Increase in Employee Productivity Due to AI (%)', 'Americans Using Voice Assistants (%)',
    'Medical Professionals Using AI for Diagnosis', 'Jobs at High Risk of Automation - Transportation & Storage (%)',
    'Jobs at High Risk of Automation - Wholesale & Retail Trade', 'Jobs at High Risk of Automation - Manufacturing'
]

# Remove % and convert to numeric
for col in percentage_columns:
    a_cleaned[col] = a_cleaned[col].str.replace('%', '').astype(float)

# Trend analysis of AI Software Revenue and Global AI Market Value
plt.figure(figsize=(10, 6))
plt.plot(a_cleaned['Year'], a_cleaned['AI Software Revenue(in Billions)'], label="AI Software Revenue (Billions)", marker='o')
plt.plot(a_cleaned['Year'], a_cleaned['Global AI Market Value(in Billions)'], label="Global AI Market Value (Billions)", marker='o')
plt.title('Trend Analysis of AI Software Revenue and Global AI Market Value')
plt.xlabel('Year')
plt.ylabel('Value in Billions')
plt.legend()
plt.grid(True)
plt.show()

# Job impact analysis: Jobs eliminated vs. created
plt.figure(figsize=(10, 6))
plt.plot(a_cleaned['Year'], a_cleaned['Estimated Jobs Eliminated by AI (millions)'], label="Jobs Eliminated by AI (Millions)", marker='o')
plt.plot(a_cleaned['Year'], a_cleaned['Estimated New Jobs Created by AI (millions)'], label="New Jobs Created by AI (Millions)", marker='o')
plt.title('Job Impact Analysis: Jobs Eliminated vs. Created by AI')
plt.xlabel('Year')
plt.ylabel('Jobs (in millions)')
plt.legend()
plt.grid(True)
plt.show()

# AI adoption and market analysis
plt.figure(figsize=(10, 6))
plt.plot(a_cleaned['Year'], a_cleaned['AI Adoption (%)'], label="AI Adoption (%)", marker='o')
plt.plot(a_cleaned['Year'], a_cleaned['Global Expectation for AI Adoption (%)'], label="Global Expectation for AI Adoption (%)", marker='o')
plt.title('AI Adoption and Market Analysis')
plt.xlabel('Year')
plt.ylabel('Adoption Percentage')
plt.legend()
plt.grid(True)
plt.show()

# AI in healthcare trend
plt.figure(figsize=(10, 6))
plt.plot(a_cleaned['Year'], a_cleaned['Medical Professionals Using AI for Diagnosis'], label="Medical Professionals Using AI for Diagnosis (%)", marker='o')
plt.plot(a_cleaned['Year'], a_cleaned['AI Contribution to Healthcare(in Billions)'], label="AI Contribution to Healthcare (Billions)", marker='o')
plt.title('AI in Healthcare: Diagnosis and Contribution')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# Jobs at high risk of automation in various sectors
plt.figure(figsize=(10, 6))
plt.plot(a_cleaned['Year'], a_cleaned['Jobs at High Risk of Automation - Transportation & Storage (%)'], label="Transportation & Storage", marker='o')
plt.plot(a_cleaned['Year'], a_cleaned['Jobs at High Risk of Automation - Wholesale & Retail Trade'], label="Wholesale & Retail Trade", marker='o')
plt.plot(a_cleaned['Year'], a_cleaned['Jobs at High Risk of Automation - Manufacturing'], label="Manufacturing", marker='o')
plt.title('Jobs at High Risk of Automation in Various Sectors')
plt.xlabel('Year')
plt.ylabel('Percentage of Jobs at Risk')
plt.legend()
plt.grid(True)
plt.show()

# AI-driven revenue increase and productivity improvements
plt.figure(figsize=(10, 6))
plt.plot(a_cleaned['Year'], a_cleaned['Estimated Revenue Increase from AI (trillions USD)'], label="Estimated Revenue Increase (Trillions USD)", marker='o')
plt.plot(a_cleaned['Year'], a_cleaned['Expected Increase in Employee Productivity Due to AI (%)'], label="Expected Employee Productivity Increase (%)", marker='o')
plt.title('Revenue and Productivity Increases Driven by AI')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# Voice assistant usage analysis
plt.figure(figsize=(10, 6))
plt.plot(a_cleaned['Year'], a_cleaned['Americans Using Voice Assistants (%)'], label="Americans Using Voice Assistants (%)", marker='o')
plt.plot(a_cleaned['Year'], a_cleaned['Digital Voice Assistants (billions of devices)'], label="Digital Voice Assistants (Billions of Devices)", marker='o')
plt.title('Voice Assistant Usage Analysis')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# AI in strategy and competitive edge analysis
plt.figure(figsize=(10, 6))
plt.plot(a_cleaned['Year'], a_cleaned['Organizations Believing AI Provides Competitive Edge'], label="Organizations Believing AI Provides Competitive Edge (%)", marker='o')
plt.plot(a_cleaned['Year'], a_cleaned['Companies Prioritizing AI in Strategy'], label="Companies Prioritizing AI in Strategy (%)", marker='o')
plt.title('AI in Strategy and Competitive Edge Analysis')
plt.xlabel('Year')
plt.ylabel('Percentage of Organizations')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Assuming a is already defined and contains the data

# Define the columns you want to plot
columns = ['Year',
 'AI Software Revenue(in Billions)',
 'Global AI Market Value(in Billions)',
 'AI Adoption (%)',
 'Organizations Using AI',
 'Organizations Planning to Implement AI',
 'Global Expectation for AI Adoption (%)',
 'Estimated Jobs Eliminated by AI (millions)',
 'Estimated New Jobs Created by AI (millions)',
 'Net Job Loss in the US',
 'Organizations Believing AI Provides Competitive Edge',
 'Companies Prioritizing AI in Strategy',
 'Estimated Revenue Increase from AI (trillions USD)',
 'Marketers Believing AI Improves Email Revenue',
 'Expected Increase in Employee Productivity Due to AI (%)',
 'Americans Using Voice Assistants (%)',
 'Digital Voice Assistants (billions of devices)',
 'Medical Professionals Using AI for Diagnosis',
 'AI Contribution to Healthcare(in Billions)',
 'Jobs at High Risk of Automation - Transportation & Storage (%)',
 'Jobs at High Risk of Automation - Wholesale & Retail Trade',
 'Jobs at High Risk of Automation - Manufacturing']

# Iterate through each column and create a bar chart or histogram
for column in columns:
    try:
        if column not in a.columns:
            print(f"Column {column} does not exist in the DataFrame")
            continue
        
        # Check if the column is suitable for a bar chart (categorical data)
        if a[column].dtype == 'object' or a[column].dtype.name == 'category':
            # Count the occurrences of each unique value
            column_counts = a[column].value_counts().reset_index()
            column_counts.columns = [column, 'count']
            
            # Create a bar chart using plotly.express
            fig = px.bar(
                column_counts,
                x=column,
                y='count',
                title=f'Distribution of {column}',
                labels={column: column, 'count': 'Count'},
                text='count'
            )
            
            # Update layout for better readability
            fig.update_layout(
                xaxis_title=column,
                yaxis_title='Count',
                paper_bgcolor='rgba(0,0,0,0)',
                plot_bgcolor='rgba(0,0,0,0)',
                title_font=dict(size=18, family="Arial"),
                xaxis={'categoryorder': 'total descending'}
            )
            
            # Show the figure
            fig.show()

        # For numerical data
        elif a[column].dtype in ['int64', 'float64']:
            # Create a histogram for numerical columns
            fig = px.histogram(
                a,
                x=column,
                title=f'Distribution of {column}',
                labels={column: column, 'count': 'Count'}
            )
            
            # Update layout for better readability
            fig.update_layout(
                xaxis_title=column,
                yaxis_title='Count',
                paper_bgcolor='rgba(0,0,0,0)',
                plot_bgcolor='rgba(0,0,0,0)',
                title_font=dict(size=18, family="Arial")
            )
            
            # Show the figure
            fig.show()
    
    except Exception as e:
        print(f"Could not create plot for column {column}: {e}")
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
from collections import Counter
import pandas as pd

# Assuming `dataset` is your DataFrame and `stop_words_list` is a list of stopwords

# Example stopwords list, modify as needed
stop_words_list = set(STOPWORDS)

# Generate frequency counts from the 'business_category' column
counts = Counter(a["Jobs at High Risk of Automation - Manufacturing"].dropna().apply(lambda x: str(x)))

# Generate the word cloud
wcc = WordCloud(
    background_color="black",
    width=1600, height=800,
    max_words=2000,
    stopwords=stop_words_list
)
wcc.generate_from_frequencies(counts)

# Display the word cloud
plt.figure(figsize=(10, 5), facecolor='k')
plt.imshow(wcc, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
